```python
import cv2
import numpy as np
import os

# Playing video from file:
cap = cv2.VideoCapture('britlong.mov')

try:
    if not os.path.exists('data'):
        os.makedirs('data')
except OSError:
    print ('Error: Creating directory of data')

currentFrame = 0
while(True):
    
    ret, frame = cap.read()

    # Saves image of the current frame in jpg file
    # path to deep dream obviously you may want to change this later
    name = '/Users/collinswestnedge/programming/jupyterProjects/deepdream/data/frame' + str(currentFrame) + '.jpg'
    print ('Creating...' + name)
    cv2.imwrite(name, frame)

    # To stop duplicate images
    currentFrame += 1
    if not ret: break

# When everything done, release the capture
cap.release()
cv2.destroyAllWindows()
```
